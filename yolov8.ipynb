```python
from ultralytics import YOLO
from roboflow import Roboflow

rf = Roboflow(api_key="")
project = rf.workspace().project("mug-detector-0akq7")
dataset = project.version(6).download("yolov8")
```

```python
model = YOLO("yolov8n.pt")  # load a pretrained model (recommended for training)

HOME = "/Users/james/src/machine-learning/Mug-Detector-6/"

# Use the model
model.train(data=HOME + "/data.yaml", epochs=3)  # train the model

metrics = model.val()

export_model = model.export(format="onnx")  # export the model to ONNX format
```

```python
import os
import cv2
import supervision as sv
from IPython.display import Image, display

HOME = "/Users/james/src/machine-learning/Mug-Detector-6/"

test_images = os.listdir(HOME + "/test/images")

all_data = []

for image in test_images:
    img = os.path.join(HOME + "/test/images", image)

    predictions = model(img)[0]  # run inference on test image

    print(predictions.boxes)

    detections = sv.Detections.from_yolov8(predictions)

    class_names = ["cup"]

    box_annotator = sv.BoxAnnotator()

    labels = [
        f"{class_names[class_id]} {confidence:0.2f}"
        for _, _, confidence, class_id, _
        in detections
    ]

    annotated_frame = box_annotator.annotate(
        scene=cv2.imread(img),
        detections=detections,
        labels=labels
    )

    all_data.append(annotated_frame)

    print(detections)

    # save
    cv2.imwrite(os.path.join("new.jpg"), annotated_frame)

display(Image(data=all_data[0], format='jpg'))
```
